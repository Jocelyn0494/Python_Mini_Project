# Decision Tree & Random Forest Demo

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load the Datasets

```python
from sklearn.datasets import make_blobs
X, Y = make_blobs(n_samples=500, centers=3,random_state=1, cluster_std=2.5)

#Scatter plot the points
plt.figure(figsize=(10,10))
plt.scatter(X[:, 0], X[:, 1], c=Y,cmap='jet')
plt.show()
```

## Decision Tree

```python
from sklearn import tree
```

```python
dtc = tree.DecisionTreeClassifier(random_state=1, max_depth=2)
dtc.fit(X,Y)

plt.figure(figsize=(10,10))
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
test_x = np.array([x1 for x1 in zip(xx.ravel(), yy.ravel())])
test_y = dtc.predict(test_x)

plt.contourf(xx, yy, test_y.reshape(xx.shape), alpha=0.3,cmap='jet')
plt.scatter(X[:,0], X[:,1], c=Y,cmap='jet')
plt.show()
```

```python
## increase the depth of the tree
dtc = tree.DecisionTreeClassifier(random_state=1, max_depth=5)
dtc.fit(X,Y)

plt.figure(figsize=(10,10))
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
test_x = np.array([x1 for x1 in zip(xx.ravel(), yy.ravel())])
test_y = dtc.predict(test_x)

plt.contourf(xx, yy, test_y.reshape(xx.shape), alpha=0.3,cmap='jet')
plt.scatter(X[:,0], X[:,1], c=Y,cmap='jet')
plt.show()
```

## Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

rtc = RandomForestClassifier(n_estimators=100,random_state=0)

rtc.fit(X,Y)

plt.figure(figsize=(10,10))
h = .02  # step size in the mesh
# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
test_x = np.array([x1 for x1 in zip(xx.ravel(), yy.ravel())])
test_y = rtc.predict(test_x)

plt.contourf(xx, yy, test_y.reshape(xx.shape), alpha=0.3,cmap='jet')
plt.scatter(X[:,0], X[:,1], c=Y,cmap='jet')
plt.show()
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
np.random.seed(1)
x = np.linspace(0, 10, 100)
noise = 0.1 * (2 * np.random.random(len(x)) - 1)
y = np.sin(x) + noise
# Plot x vs y
plt.figure(figsize=(16,8))
plt.scatter(x, y)
plt.show()
```

```python
rtr = RandomForestRegressor(n_estimators=100,random_state=0)
rtr.fit(x.reshape(-1,1),y)
y_predict = rtr.predict(x.reshape(-1,1))

# Plot 
plt.figure(figsize=(16,8))
plt.scatter(x, y)

plt.plot(x, y_predict, '-r');
plt.plot(x, np.sin(x), '-k', alpha=0.5);
```
