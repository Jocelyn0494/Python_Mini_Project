# Numpy where function

```python
A =np.array([1,2,3,4])
```

```python
B = np.array([5,1,7,2])
```

```python
condition = np.array([True,False,True,False])
```

```python
np.where(condition,A,B) #True is returning A, else return B
```

```python
b = np.random.randn(5,5)
```

```python
b
```

```python
np.where(b<0,0,b)
```
