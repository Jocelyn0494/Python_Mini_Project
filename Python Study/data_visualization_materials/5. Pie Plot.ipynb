```python
import matplotlib.pyplot as plt
```

```python
label = 'Cats','Birds','Dogs','Pandas'
```

```python
size = [25,35,30,10] #percentage
```

```python
explode = (0,0.1,0.2,0)#每一块饼之间联系另一块饼的紧密程度的参数
```

```python
fig, ax = plt.subplots()
ax.pie(size,explode = explode,labels = label,autopct='%.2f %%',shadow=False,startangle=45)#shadow:让图中具有一定的阴影；startangle: 第一个pie chart 的角度
ax.axis('equal')#调节整个的图是完美的圆
plt.show()
```
