```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15*np.random.rand(N)) ** 2 #specify 每个点的大小
```

```python
plt.scatter(x,y,c=colors,s=area,alpha=0.5)#alpha=0.5 透明度
```

```python
plt.show()
```
