```python
import numpy as np
import matplotlib.pyplot as plt
```

# Twin Axes

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

ax2 = ax1.twinx()

t = np.linspace(0.,10.,100)
ax1.plot(t,t**2,'b-')
ax1.set_ylabel('Density(cgs)',color='red')

ax2.plot(t,1000/(t+1),'r-')
ax2.set_ylabel('Temperature',color='blue')

ax1.set_xlabel('Time')

plt.show()
```

# Add Legend

```python
fig = plt.figure()# 定义一个图像
ax = fig.add_subplot(1,1,1) #定义子图的画图区间，画布

x = np.linspace(1.,8.,30)#从1 到 8 里面有30 个点
ax.plot(x,x**1.5,'ro',label='density')# ro： 红点的形式
ax.plot(x,20/x,'bx',label='temperatures')# bx: 蓝x

# add legend 把右上角的density 和 temperatures 加上去
ax.legend(loc='upper right')
plt.show()
```

# Add a Colorbar

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

image = np.random.poisson(10.,(100,80))#poisson 分布 # image 是一个（100行，80列）的numpy array.
i = ax.imshow(image)

# adjust the position of color bar
colorbar_ax = fig.add_axes([1,0,0.05,1])#1:距离图左侧的距离，图左侧的位置； 0：调整上下轴的位置； 1：bar的宽度；1：color bar 的高度

fig.colorbar(i,cax=colorbar_ax)
plt.show()
```

# Custom ticks

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1)

x = [1,2,3,4,5,6,7]
y = [4,3,2,3,4,5,4]

ticks = ['O','B','A','F','G','K','M']

# Plot our data
ax.plot(x,y,'ro')

# set the limits
ax.set_xlim(0.5,7.5)
ax.set_ylim(0.,10.)

# set the custom ticks on x-axis
ax.set_xticks(x) #定义ticks 的位置
ax.set_xticklabels(ticks)

# set axis labels
ax.set_xlabel('x value')
ax.set_ylabel('y value')


plt.show()

```
